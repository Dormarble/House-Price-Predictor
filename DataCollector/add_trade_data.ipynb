```python
import pandas as pd
import numpy as np
```

## Data 로드

```python
trade_data_df = pd.read_csv('./data/trade_data.csv')
```

# 1. 병원 데이터 반영

```python
hospitalFile = open('./result/id_dist_hospital.txt', 'r', encoding='ANSI')
hospitalFile = hospitalFile.read().split('\n')[:-1]

hospitalDist = {}
for content in hospitalFile:
    content = content.split(' ')
    hospitalDist[int(content[0])] = float(content[1])
print(len(hospitalDist))
```

```python
trade_data_df['distHospital'] = trade_data_df.apply(lambda x: hospitalDist[x['id']], axis=1)
```

```python
trade_data_df
```
